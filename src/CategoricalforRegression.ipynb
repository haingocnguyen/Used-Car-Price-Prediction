```python
import pandas as pd
# from sklearn.linear_model import LinearRegression 
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = 'data\CarPrice_Assignment.xlsx'
df = pd.read_excel(file_path, sheet_name="CarPrice_Assignment", header=0)
```

### **Regression for cylinder's attribute**

```python
df_dummies = pd.read_csv('data\dummy_data.csv')
```

```python
df_dummies
```

```python
df['cylindernumber']
```

```python
df['cylindernumber'].unique()
```

```python
from patsy import dmatrices
```

```python
y, X = dmatrices('price ~ C(cylindernumber)', data=df, return_type='dataframe')

# Fit the OLS model
model = sm.OLS(y, X)
reg = model.fit()

# Print the summary of the regression
print(reg.summary())
```

### **Regression for 6 dummies of cylindernumber**

#### Splitting in to a separated dataframe

```python
df_cylinder = df_dummies[['cylindernumber_five', 'cylindernumber_four', 'cylindernumber_six', 'cylindernumber_three', 'cylindernumber_twelve', 'cylindernumber_two']]
df_cylinder
```

```python
# Columns to be label encoded
columns_to_encode = ['cylindernumber_five', 'cylindernumber_four', 'cylindernumber_six', 'cylindernumber_three', 'cylindernumber_twelve', 'cylindernumber_two']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to each column
for column in columns_to_encode:
    df_cylinder[column] = label_encoder.fit_transform(df_cylinder[column])

# Display the transformed DataFrame
df_cylinder
```

```python
df_cylinder['price'] = df_dummies['price1']
```

```python
df_cylinder
```

```python
df_cylinder.corr()
```

#### OLS Summary

```python
y, X = dmatrices('price ~ C(cylindernumber_two)', data=df_cylinder, return_type='dataframe')

# Fit the OLS model
model = sm.OLS(y, X)
reg = model.fit()

# Print the summary of the regression
print(reg.summary())
```
