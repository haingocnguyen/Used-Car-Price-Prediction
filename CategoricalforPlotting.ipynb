```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'data\CarPrice_Assignment.xlsx'
df = pd.read_excel(file_path, sheet_name="CylinderNumber", header=0)
df
```

```python
#Frequency Bar Graph
print("Frequency Bar Graph")
plt.bar(df.CylinderNumber, df.Frequency, width=0.4)
plt.xlabel("Cylinder Number")
plt.ylabel("Frequency")
plt.title("Bar Graph for Cars' Cylinder Number Frequency")
# plt.xticks(rotation=90)
plt.show()
```

```python
#Relative Creation
df['Relative_Frequency'] = None
for i in range(7):
    df['Relative_Frequency'] = df['Frequency']/sum(df['Frequency'])
print("Relative Frequency Table")
print(df)
```

```python
#Relative Frequency Bar Graph
print("Relative Frequency Bar graph")
plt.bar(df.CylinderNumber, df.Relative_Frequency, width = 0.4)
plt.xlabel("Cylinder Number")
plt.ylabel("Relative Frequency")
plt.title("Bar Graph for Cars' Cylinder Number Relative Frequency")
# plt.xticks(rotation=90)
plt.show()
```

```python
#Pie chart for Frequency
print("Frequency Pie Chart")
plt.figure(figsize=(6, 8))
plt.pie(df['Frequency'], labels=df['CylinderNumber'], autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart for Cars' Cylinder Number Frequency")
plt.show()
```

```python
#Pie Chart for Relative Frequency
print("Relative Frequency Pie Chart")
plt.figure(figsize=(6, 8))
plt.pie(df['Relative_Frequency'], labels=df['CylinderNumber'], autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart for Cars' Cylinder Number Relative Frequency")
plt.show()
```
